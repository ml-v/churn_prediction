```python
import pandas as pd 
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix

pd.set_option('future.no_silent_downcasting', True)
```

```python
data = pd.read_csv('data.csv', sep=',', decimal=',')
data
```

```python
-(len(data.dropna()) - len(data))
```

```python
data[data.isna().any(axis=1)]
```

```python
data = data.dropna().drop(labels=['customerID', 'TotalCharges'], axis=1)
```

```python
data.info()
```

```python
for column in data.columns:
    if column not in ['SeniorCitizen', 'tenure', 'MonthlyCharges']:
        print(f'"{column}": {len(data[column].unique())} || {list(data[column].unique())}')
```

```python
mappings = {
    'gender': {'Female': '0', 'Male': '1'},
    'Partner': {'No': '0', 'Yes': '1'},
    'Dependents': {'No': '0', 'Yes': '1'},
    'PhoneService': {'No': '0', 'Yes': '1'},
    'MultipleLines': {'No phone service': '0', 'No': '1', 'Yes': '2'},
    'InternetService': {'No': '0', 'Fiber optic': '1', 'DSL': '2'},
    'OnlineSecurity': {'No internet service': '0', 'No': '1', 'Yes': '2'},
    'OnlineBackup': {'No internet service': '0', 'No': '1', 'Yes': '2'},
    'DeviceProtection': {'No internet service': '0', 'No': '1', 'Yes': '2'},
    'TechSupport': {'No internet service': '0', 'No': '1', 'Yes': '2'},
    'StreamingTV': {'No internet service': '0', 'No': '1', 'Yes': '2'},
    'StreamingMovies': {'No internet service': '0', 'No': '1', 'Yes': '2'},
    'Contract': {'Month-to-month': '0', 'One year': '1', 'Two year': '2'},
    'PaperlessBilling':  {'No': '0', 'Yes': '1'},
    'PaymentMethod': {'Bank transfer (automatic)': '0', 'Electronic check': '1', 
                      'Mailed check': '2', 'Credit card (automatic)': '3'},
    'Churn':  {'No': '0', 'Yes': '1'}
}
```

```python
model_data = data.replace(mappings)
model_data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(model_data.drop('Churn', axis=1), model_data['Churn'], 
                                                    test_size=0.3, random_state=12)
```

```python
gs_params = {'n_estimators': [70, 80, 90, 100, 110, 120, 130], 
             'min_samples_split': [2, 3, 4, 5],
             'criterion': ['gini', 'entropy'],
             'bootstrap': [True]}
rfc = RandomForestClassifier(random_state=123)
clf = GridSearchCV(rfc, gs_params, verbose=3)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
rfc = RandomForestClassifier(n_estimators=110, criterion='gini', min_samples_split=5, 
                             bootstrap = True, random_state=123)
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
print(f'Train score: {round(rfc.score(X_train, y_train) * 100, 2)}%')
print(f'Prediction accuracy: {round(accuracy_score(y_test, y_pred) * 100, 2)}%')
print(f'Confusion matrix: \n {pd.DataFrame(confusion_matrix(y_test, y_pred, labels=['0', '1']))}')
```

```python
features_importance = pd.DataFrame({'features': rfc.feature_names_in_, 
              'importance': rfc.feature_importances_}).sort_values('importance',ascending=False)
features_importance
```
